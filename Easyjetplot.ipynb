```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ejt = pd.read_excel('/Stockdataset.xlsx', engine = 'openpyxl', sheet_name = 'Easyjet')
```

```python
ejt['Date'] = pd.to_datetime(ejt['Date'])
ejt.set_index('Date', inplace=True)
ejt.sort_index(inplace=True)
```

```python
plt.figure(figsize = (12,6))
plt.plot(ejt['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Easyjet Stock Price')
plt.show()
```
